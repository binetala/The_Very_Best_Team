```python
#Dependencies
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress
import time
#import json
#import requests
import quandl
```

```python
zillow_path = "zillow_ke.csv"
zillow_df = pd.read_csv(zillow_path)
zillow_df = zillow_df.rename(columns={"Date" : "Date Zillow"})
zillow_df.head()
```

```python
unemployment_path = "unemployment_ke.csv"
unemployment_df = unemployment_df.dropna(how="any")
unemployment_df = pd.read_csv(unemployment_path)
unemployment_df = unemployment_df.rename(columns={"Year" : "Year Unemployment", "Period": "Month Unemployment"})
unemployment_df.head()
```

```python
unemployment_df["TimeTwo"] = unemployment_df["Year Unemployment"].astype(str) + "-" + unemployment_df["Month"].astype(str)
unemployment_df.head()
```

```python
interest_path = "interest_ke.csv"
interest_df = pd.read_csv(interest_path)
interest_df = interest_df.rename(columns={"Date" : "Date Interest"})
interest_df = interest_df.dropna(how = "any")
interest_df.head()
```

```python
interest_df["Date Interest"].dtype
```

```python
interest_df["Date Interest"] = interest_df["Date Interest"].astype(str)
interest_df["Date Interest"].dtype
```

```python
# interest_df['Date Interest'] = interest_df['Date Interest'].apply(pd.to_datetime)
# interest_df['Date Interest'] = pd.to_datetime(interest_df['Date Interest'], format = '%Y%B')
# interest_df['Date Interest'].dtype
```

```python
interest_df["Month"] = interest_df["Date Interest"].str.split("/")[0]
interest_df.head()
```

```python
interest_df.dtype
```

```python
stage1 = pd.merge(zillow_df, unemployment_df, how='outer', on='Timepoint')
stage1.head()
```

```python
stage2 = pd.merge(stage1, interest_df, how='outer', on='Timepoint')
stage2.head(2)
```

```python
matchmaker_df = stage2[["Year Unemployment", "Month Unemployment", "Timepoint", "Region ID", "Region Name", "Region Type", "Indicator ID", "Indicator", "Value", "unemployment rate", "Wells Fargo Rate", "Freddie Mac Rate", "Date Zillow" , "Date Interest"]]
matchmaker_df.head()
```

```python
mixmaster_df = matchmaker_df.rename(columns={"Indicator" : "Indicator Name", "unemployment rate":"Unemployment Rate"})
mixmaster_df.head()
```

```python
mixmaster_df.to_csv("mixmaster.csv", index=False, header=True)
```
